```python
print(train['Length'].skew())
train["Length"] = (train["Length"]) ** (2.25) 

print(train['Length'].skew())

print(train['Length'].unique())
```

```python
from sklearn.preprocessing import FunctionTransformer, StandardScaler
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.compose import make_column_selector, make_column_transformer



power_func = make_pipeline(
    FunctionTransformer(lambda x: (x+3) ** 2.25),
    StandardScaler(),
)

sq_func = make_pipeline(
    FunctionTransformer(lambda x: (x+ 3) ** (1/1.5)),
    StandardScaler(),
)

default = make_pipeline(
    StandardScaler()
)

from sklearn.preprocessing import OrdinalEncoder

columns = make_column_transformer(
    (power_func, ["Length"]),
    (sq_func, ['Whole weight', 'Diameter', 'Height',
       'Whole weight.1', 'Whole weight.2', 'Shell weight']),
    (OrdinalEncoder(), ['Sex']),
     verbose_feature_names_out=False, 
     remainder='passthrough' 
)

from sklearn.linear_model import SGDRegressor


pipeline = make_pipeline(columns, RandomForestClassifier())
```

Data after transformation

```python
Y_target = train['Rings']
X_data = train.drop("Rings", axis=1)

pipeline.fit(X_data, Y_target)
```

```python
from sklearn.metrics import mean_squared_log_error
y_pred = pipeline.predict(X_data)


print(f'Log RSME is {mean_squared_log_error(y_pred, Y_target)}')
```

```python



submission = pd.DataFrame()

submission['id'] = test['id']

submission['Rings'] = pipeline.predict(test)

submission.to_csv("abalone.csv", header=True, index=False)

```
