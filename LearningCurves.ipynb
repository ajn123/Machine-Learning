# Learning Curves

Learning Curves can help you if you are underfitting
- The model is converging too quickly and the error is too high

or overfitting

- The data is performing vastly better on the training data then the validation test

```python
from sklearn.model_selection import learning_curve, LearningCurveDisplay
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder, LabelEncoder
import seaborn as sns
import numpy as np

tips = sns.load_dataset("iris")

from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.linear_model import SGDClassifier, LogisticRegression, RANSACRegressor
from sklearn.ensemble import RandomForestClassifier

y = tips['species']
x = tips.drop('species', axis=1)


y = LabelEncoder().fit_transform(y)

pipeline = make_pipeline(
                StandardScaler(), 
                LogisticRegression()
                         )

train_size, train_scores, test_scores = learning_curve(pipeline, x,y, 
                                                       cv=5, shuffle=True,
                                                       scoring="neg_root_mean_squared_error")

display = LearningCurveDisplay(train_sizes=train_size, train_scores=train_scores, test_scores=test_scores, score_name="neg_root_mean_squared_error")

display.plot()

```
