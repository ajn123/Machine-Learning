# Convolutional Nueral Network

- Remember to have a reshape layer to convert the columns into an image
- A Rescale layer to keep all values between 0 and 1

```python
from tensorflow import keras



model = keras.Sequential(
    [
        keras.layers.Reshape((28, 28, 1)),
        keras.layers.Rescaling(1.0/255),
        DefaultConv2d(filters=64, kernel_size=7, input_shape=(28, 28, 1)),
        keras.layers.MaxPool2D(),
        DefaultConv2d(filters=128),
        DefaultConv2d(filters=128),
        keras.layers.MaxPool2D(),
        DefaultConv2d(filters=256),
        DefaultConv2d(filters=256),
        keras.layers.MaxPool2D(),
         DefaultConv2d(filters=512),
        DefaultConv2d(filters=512),
        keras.layers.MaxPool2D(),
        keras.layers.Flatten(),
        keras.layers.Dense(300, activation='relu'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(300, activation='relu'),
        keras.layers.Dropout(0.5),
            keras.layers.Dense(300, activation='relu'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(300, activation='relu'),
        keras.layers.Dropout(0.5),
        keras.layers.Dense(10,  activation='softmax')   
    ]
)
```

- Use 'categorical_crossentropy' because we one hot encoded the values above

```python
from tensorflow.keras.optimizers import SGD

sgd = SGD(momentum=0.9, learning_rate=0.01)

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X, y_cat, validation_split=0.2, epochs=5)
```
