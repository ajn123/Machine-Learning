```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



train = pd.read_csv('./datasets/house-prices-advanced-regression-techniques/train.csv')
test = pd.read_csv('./datasets/house-prices-advanced-regression-techniques/test.csv')


train
```

```python

def transform_features(df):
    df = (
    df.assign(YrSold=lambda x: x.YrSold.astype('category'),
              )
    )
    return df



transform_features(train)
    
```

```python
def encode_columns(df):
    one_hot_encoded_columns = []
    for col in df.columns:
        if len(df[col].unique()) <= 5:
            one_hot_encoded_columns.append(col)
            
    return one_hot_encoded_columns
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

from category_encoders import TargetEncoder

import sklearn
sklearn.set_config(transform_output="pandas")


y = train['SalePrice']
X = train.drop('SalePrice', axis=1)

#one_hot_encoded_columns= one_hot_encoded_columns.remove("SalePrice")

column_trans = make_column_transformer(
    (TargetEncoder(), make_column_selector(dtype_exclude=[np.number])),
    verbose_feature_names_out=False, remainder='passthrough'
)

object_trans = make_column_transformer(
    (OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), make_column_selector(dtype_include=object)),
    verbose_feature_names_out=False, remainder='passthrough'
)

imputer = make_column_transformer(
    (SimpleImputer(strategy="most_frequent"), make_column_selector(dtype_include=[np.generic])),
     remainder='passthrough', verbose_feature_names_out=False
)

pipe = make_pipeline(column_trans, object_trans, imputer,)



from sklearn.model_selection import cross_val_score


cross_val_score(make_pipeline(pipe, LinearRegression()), X,y, cv=3, scoring='neg_root_mean_squared_error')



pipe.fit_transform(X,y).shape
```

Let's try rigdge Regression

```python

params = {'ridge__alpha':  np.linspace(25,35,40)}


from sklearn.linear_model import Ridge
from sklearn.model_selection import RandomizedSearchCV

ridge = Ridge()
ridge = make_pipeline(pipe, ridge)

rscv = RandomizedSearchCV(
    ridge,
    params,
    cv=4, 
    scoring='neg_root_mean_squared_error',
    return_train_score=True,
    refit=True
)


rscv.fit(X,y)
print(rscv.best_score_)
print(rscv.best_params_)
```

Let's try XGBoost

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, shuffle=True, random_state=42)

#cross_val_score(make_pipeline(pipe, XGBRegressor(n_estimators=1000)), X,y, cv=3, scoring="neg_root_mean_squared_error", verbose=True)


xgboostr = XGBRegressor(n_estimators = 1000, learning_rate=.05)

pipe = make_pipeline(pipe, xgboostr)
model = pipe.fit(X_train,y_train)


```

```python
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestRegressor


selector = RFECV(RandomForestRegressor(), step=1, cv=5, scoring="neg_root_mean_squared_error", min_features_to_select=15)


X = pipe.fit_transform(X,y)
selector.fit(X,y)

print(selector.support_)
```
